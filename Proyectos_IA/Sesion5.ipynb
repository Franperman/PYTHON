```python
import numpy as np
import pandas as pd

asignaturas = ["matemáticas", "física", "historia"]
notas = [9, 9, 8]
clases = pd.Series(data=notas, index=asignaturas)

nota_fisica = clases["física"]

print(f"La nota de física es un: {nota_fisica}")

print("La nota de física es un: {}".format(nota_fisica))
```

```python
ventas_enero = list(np.random.randint(100, 501, 3))
ventas_febrero = list(np.random.randint(100, 501, 3))
ventas_marzo = list(np.random.randint(100, 501, 3))
filas = ["ventas_enero", "ventas_febrero", "ventas_marzo"]
columnas = ["oficina1", "oficina2", "oficina3"]
ventas = [ventas_enero, ventas_febrero, ventas_marzo]
dataframe = pd.DataFrame(data=ventas, columns=columnas, index=filas)
print(dataframe)

print("Las ventas de enero para la oficina 1 son de: {}".format(dataframe["oficina1"]["ventas_enero"]))
```

```python
import pandas as pd

def calcular_promedio(dataframe):
    medio = dataframe.mean()
    return medio

# Ejemplo de uso
data = {
        'A': [1, 2, 3, 4],
        'B': [5, 6, 7, 8],
        'C': [9, 10, 11, 12]
    }
df = pd.DataFrame(data)
resultado = calcular_promedio(df)
print(resultado)

```

```python
import pandas as pd

def seleccionar_datos(dataframe, criterio):
    dato = dataframe.query(criterio)
    return dato
    

# Ejemplo de uso
data = {
        'nombre': ['Alice', 'Bob', 'Charlie', 'David'],
        'edad': [20, 22, 18, 25],
        'calificaciones': [90, 88, 75, 95]
    }

df = pd.DataFrame(data)
criterio = 'edad > 18'
resultado = seleccionar_datos(df, criterio)
print(resultado)
```

```python
import pandas as pd

def rellenar_con_media(dataframe, columna):
    dataframe[columna] = dataframe[columna].fillna(dataframe[columna].mean())
    return dataframe


# Ejemplo de uso
data = {
        'nombre': ['Alice', 'Marta', 'Charlie', 'David', 'Eve'],
        'edad': [20, None, 18, 25, None],
        'calificaciones': [90, 88, None, None, 95]
    }

df = pd.DataFrame(data)
columna = 'calificaciones'

resultado = rellenar_con_media(df, columna)
print(resultado)
```

```python
import matplotlib.pyplot as plt



x = np.linspace(2, 9, 10)

y = x ** 2

plt.plot(x,y, "b")
plt.title("grafico")
plt.ylabel("etiqueta")
plt.xlabel("peñiscola")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"g")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.9,0.8])
grafico.plot(x,y,"g")
```

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x,y,"b")

grafico2 = figura.add_axes([0.15,0.68,0.3,0.3])
grafico2.plot(y,x,"y")
```

```python
figura3 = plt.figure()
graficos1 = figura3.add_axes([0.1, 0.1, 0.9, 0.9])
graficos1.plot(x, x ** 2, label = "Al cuadrado")
graficos1.plot(x, x ** 3, label = "Al cubo")
graficos1.legend(loc=0)
```
