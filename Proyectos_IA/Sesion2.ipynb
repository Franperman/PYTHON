```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Función para generar datos que siguen una distribución normal
def generar_datos_gaussianos(media, desviacion, cantidad):
    datos = [random.gauss(media, desviacion) for _ in range(cantidad)]
    return datos

# Parámetros de la distribución
media = 0  # Media de la distribución
desviacion = 1  # Desviación estándar de la distribución
cantidad = 1000  # Cantidad de datos a generar

# Generar datos
valores = generar_datos_gaussianos(media, desviacion, cantidad)

# Visualizar la distribución con un histograma
plt.hist(valores, bins=50, density=True, alpha=0.6, color='b')

# Crear la curva de la Campana de Gauss para comparar
x = np.linspace(min(valores), max(valores), 100)
curva_gauss = (1 / (desviacion * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - media) / desviacion) ** 2)
plt.plot(x, curva_gauss, color='r')

# Configuración del gráfico
plt.title('Campana de Gauss - Distribución Normal')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los parámetros
media = 0  # Media de la distribución
desviacion = 1  # Desviación estándar
cantidad = 10000  # Cantidad de datos a generar
iteraciones = 12  # Número de iteraciones para simular una distribución normal

# Inicializar lista de valores
valores = []

# Generar datos que sigan una distribución normal usando sumas aleatorias
for _ in range(cantidad):
    suma = 0
    for _ in range(iteraciones):
        suma += np.random.uniform(-0.5, 0.5)
    valor = media + desviacion * suma
    valores.append(valor)

# Convertir la lista de valores a un array de numpy
valores = np.array(valores)

# Graficar el histograma
plt.hist(valores, bins=50, density=True, alpha=0.6, color='b')

# Crear la curva de la Campana de Gauss para comparar
x = np.linspace(min(valores), max(valores), 100)
curva_gauss = (1 / (desviacion * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - media) / desviacion) ** 2)
plt.plot(x, curva_gauss, color='r')

# Configuración del gráfico
plt.title('Aproximación a la Campana de Gauss - Distribución Normal')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show()
```
