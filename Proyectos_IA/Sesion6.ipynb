```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
casas.head()
```

```python
x = casas[["Ingreso_rea", "Antigedad_rea", "Superficie_media/habitaciones", "Superficie_media/dormitorios", "Poblacin_zona"]]
y = casas["Precio"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
lrm = LinearRegression()
lrm.fit(x_train, y_train)
```

```python
predicciones = lrm.predict(x_test)
plt.scatter(y_test, predicciones)
```

```python
sns.histplot(y_test - predicciones)
```

<span style="font-size:30px;">MAE (Mean absolute error) - Media del valor absoluto de los errores</span>

```python
metrics.mean_absolute_error(y_test, predicciones)
```

<span style="font-size:30px;">MSE (Media de los errores al cuadrado)</span>

```python
np.sqrt(metrics.mean_squared_error(y_test, predicciones))
```

```python
# Funcin de regresin lineal
def regresion_ventas(datos):
    x = datos[['TV', 'Radio', 'Periodico']]
    y = datos['Ventas']

    lrm = LinearRegression()
    lrm.fit(x, y)
    return lrm



# Ejemplo de uso con datos reales
data = {
    'TV': [230.1, 44.5, 17.2, 151.5, 180.8],
    'Radio': [37.8, 39.3, 45.9, 41.3, 10.8],
    'Periodico': [69.2, 45.1, 69.3, 58.5, 58.4],
    'Ventas': [22.1, 10.4, 9.3, 18.5, 12.9]
}
df = pd.DataFrame(data)
modelo_regresion = regresion_ventas(df)


# Estimaciones de ventas para nuevos datos de inversin en publicidad
nuevos_datos = pd.DataFrame({'TV': [200, 60, 30], 'Radio': [40, 20, 10], 'Periodico': [50, 10, 5]})
estimaciones_ventas = modelo_regresion.predict(nuevos_datos)

print("Estimaciones de Ventas:")
print(estimaciones_ventas)

```

```python
def edad_media(columnas):
    edad = columnas[0]
    clase = columnas[1]
    if pd.isnull(edad):
        if clase == 1:
            return 38
        if clase == 2:
            return 30
        else:
            return 25
    else:
        return edad
```

```python
entrenamiento["Embarcado"].fillna("S", inplace=True)
```

```python
entrenamiento["Edad"] = entrenamiento[["Edad", "Clase"]].apply(edad_media, axis=1)

entrenamiento.drop(["Nombre", "Cabina", "Boleto", "Identificacin"], axis=1, inplace=True)

sns.heatmap(entrenamiento.isnull())
```

```python
entrenamiento.info()
```

```python
sns.countplot(x="Embarcado", data=entrenamiento, hue= "Clase")
```

```python
sexo = pd.get_dummies(entrenamiento["Sexo"], drop_first=True, dtype = int)

sexo.columns = ["Sexo"]

entrenamiento = pd.concat([entrenamiento, sexo], axis=1)
sexo
```

```python
entrenamiento.columns = ["Sobrevivi", "Clase", "pedro", "Edad", "Familiar", "Abrasar", "Tarifa", "Embarcado", "Sexo"]
entrenamiento.drop("pedro", inplace=True, axis=1)
entrenamiento
```

```python
def durmiente(dormir):
        if dormir == "C":
                return 0
        elif dormir== "Q":
                return 1
        elif dormir== "S":
                return 2
        else:
                return dormir

entrenamiento["Embarcado"] = entrenamiento["Embarcado"].apply(durmiente)

entrenamiento
```

```python
yy = entrenamiento["Sobrevivi"]
xx = entrenamiento[["Clase", "Edad", "Familiar", "Abrasar", "Tarifa", "Embarcado", "Sexo"]]
yy.head()
```
