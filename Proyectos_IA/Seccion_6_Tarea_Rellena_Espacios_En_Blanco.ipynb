```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Importa los datos
peliculas = pd.read_csv('C:\\Users\\Fanper\\Documents\\PYTHON\Proyectos_IA\\Seccion_6_Tarea_Datos.csv', encoding = 'latin1')
```

```python
# Explora la variable categórica Género, usada en el reto
peliculas.Genero.value_counts()
```

```python
# Filtra el DataFrame por Género
peliculas2 = peliculas[(peliculas.Genero == 'accion') | (peliculas.Genero == 'aventura') | (peliculas.Genero == 'animacion') | (peliculas.Genero == 'comedia') | (peliculas.Genero == 'drama')]
```

```python
# Filtra el DataFrame por Estudio
peliculas3 = peliculas2[(peliculas2.Estudio == 'Buena Vista Studios') | (peliculas2.Estudio == 'Fox') | (peliculas2.Estudio == 'Paramount Pictures') | (peliculas2.Estudio == 'Sony') | (peliculas2.Estudio == 'Universal') | (peliculas2.Estudio == 'WB')]
```

```python
# Revisa que los filtros se hayan aplicado
print (peliculas2.Genero.unique())
print (peliculas3.Estudio.unique())
print (len(peliculas))
```

```python
# Define el estilo de los gráficos
sns.set_theme(style="darkgrid", palette="muted", color_codes=True)

# Gráfica los boxplots para visualizar la distribución de los ingresos por género
ax = sns.boxplot(data=peliculas2, x='Genero', y='Ingresos % USA', orient='v', color='lightgray', showfliers=False)

# Ajusta la transparencia de los cuadros de los boxplots
plt.setp(ax.artists, alpha=0.5)

# Agrega los puntos individuales para mostrar cada observación, añadiendo un efecto de dispersión (jitter)
sns.stripplot(x='Genero', y='Ingresos % USA', data=peliculas3, jitter=True, size=6, linewidth=0, hue='Estudio', alpha=0.7)

# Configura el título y las etiquetas de los ejes
ax.axes.set_title('Ingresos Brutos (%) Domésticos por Género', fontsize=30)
ax.set_xlabel('Genero', fontsize=20)
ax.set_ylabel('Ingreso Bruto (%) USA', fontsize=20)

# Define la posición de la leyenda fuera del gráfico para evitar solapamientos
ax.legend(bbox_to_anchor=(1.05, 1), loc=2)

# Muestra el gráfico
plt.show()
```
