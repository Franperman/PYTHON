```python
import pandas as pd

# Load the Excel file
file_path = 'ETC Rolling Bid Summary V3.xlsx'
xls = pd.ExcelFile(file_path)

# Define a function to clean and normalize data
def clean_normalize(df):
    # Transpose DataFrame to switch rows and columns
    df_transposed = df.transpose()
    
    # Fill NaN values with empty string or any other appropriate value
    df_cleaned = df_transposed.fillna('')
    
    # Normalize data if needed
    # For example, you might want to convert string numbers to numeric values
    # df_cleaned['YourColumn'] = pd.to_numeric(df_cleaned['YourColumn'], errors='coerce')
    
    return df_cleaned

# Load each sheet into a DataFrame, clean and normalize the data
df_public_side = clean_normalize(pd.read_excel(xls, 'Public Side Bid Tracker'))
df_119488 = clean_normalize(pd.read_excel(xls, '119488'))
df_119409 = clean_normalize(pd.read_excel(xls, '119409'))
df_sign_shop = clean_normalize(pd.read_excel(xls, 'Sign Shop Inventory'))

# Display a summary of the data
print(df_public_side.head())
print(df_119488.head())
print(df_119409.head())
print(df_sign_shop.head())

# Define file paths for saving the cleaned DataFrames
csv_file_public_side = 'C:/Proyectos/ETC_public_side.xlsx'
csv_file_119488 = 'C:/Proyectos/ETC_119488.xlsx'
csv_file_119409 = 'C:/Proyectos/ETC_119409.xlsx'
csv_file_sign_shop = 'C:/Proyectos/ETC_sign_shop.xlsx'

# Save the cleaned DataFrames to Excel files
df_public_side.to_excel(csv_file_public_side, index=False)
df_119488.to_excel(csv_file_119488, index=False)
df_119409.to_excel(csv_file_119409, index=False)
df_sign_shop.to_excel(csv_file_sign_shop, index=False)
```
